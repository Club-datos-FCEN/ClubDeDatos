```python
import pandas as pd

data = pd.read_csv("GlobalTemperatures.csv")
display(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style()
data['dt_datetime'] = pd.to_datetime(data['dt'])
data['year'] = data['dt_datetime'].dt.year
filtrado = data[data['dt_datetime'].dt.year >= 1940]
```

```python
filtrado.plot(x='dt_datetime', y='LandAverageTemperature')
```

```python
anuales = filtrado.groupby(by='year').mean()
display(anuales)
anuales.plot(y='LandAverageTemperature')
```
