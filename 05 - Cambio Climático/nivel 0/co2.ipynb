```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style()
sns.set_theme()
```

```python
data = pd.read_csv('annual-co-emissions-by-region.csv').rename(columns={'Annual CO2 emissions (zero filled)':'Emisiones'})
display(data)
```

```python
anuales = data.groupby(by='Year').sum()
anuales['Emisiones'].plot(figsize=(12,8))
plt.ylabel('Emisiones CO2 (miles de millones de Toneladas)')
plt.xlabel('AÃ±o')
plt.title('Emisiones Mundiales de Carbono', size=20)
plt.savefig('int-co2.jpg')
plt.show()
```
